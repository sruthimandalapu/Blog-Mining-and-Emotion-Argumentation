```python
import csv
import random
import numpy as np 
import matplotlib.pyplot as plt
lst =[]
before=[]
after=[]
title=[]
i=0
while(i<5):
	n = random.randint(1,15)
	if(n in lst):
		continue
	lst.append(n)
	i+=1
with open('analyze_details.csv',encoding="utf8") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
    	if(row[0].isdigit()):
	    	if(int(row[0]) in lst):
	    		before.append(int(row[2]))
	    		after.append(int(row[3]))
	    		title.append("Blog"+str(row[0]))
X_axis = np.arange(len(before))
plt.bar(X_axis-0.2,before,0.4,label = 'Before Summarization')
plt.bar(X_axis+0.2,after,0.4,label = 'After Summarization')
plt.xticks(X_axis, title,fontsize=11,fontname="Calibri")     #red and green
plt.yticks(fontsize=11,fontname="Calibri")
plt.xlabel("Blog Posts",fontsize=11,fontname="Calibri",labelpad=8)
plt.ylabel("Number of Lines of Text",fontsize=11,fontname="Calibri",labelpad=8)
plt.title("Lines of Text of a Blog Before and After Sumarization (Dynamic)",fontsize=11,fontname="Calibri",pad=15)
plt.legend(fontsize=7,borderpad=1.2,prop={'family':'Calibri', 'size':10})
plt.show()
```
